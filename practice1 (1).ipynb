```python
#Write a program to accept working hours from user and check with the working hours is 
#greater than 8 if yes give bonus of 1000 to the employee or else not.

working_hours=int(input("Enter the working_hours"))
working_hours=5000
bonus=1000

if working_hours>8:
    salary=working_hours+bonus
    print(salary)
    
```

```python
#Write a program to accept two numbers and find the largest number or if they are equal
num1=int(input("Enter a number"))
num2=int(input("Enter a number"))

if num1>num2:
    print(f"{num1} is greater than {num2}")
elif num2>num1:
    print(f"{num2} is greater than {num1}")
elif num1==num2:
    print(f"{num1} is equal to {num2}")
```

```python
#write a program to accept marks of three subjects from user.
#calculate the total marks for all three subjects 
#calculate the percentage secured my the student
#based on the percentage give the student his grade 
# if   pass and fail
#    percentage should be above 65 -- > Pass else fail

maths=int(input("Enter a marks for math"))
english=int(input("Enter a marks for english"))
science=int(input("Enter a marks for science"))
print("........marks obtained..........")
print(f"maths :{maths}")
print(f"english :{english}")
print(f"science: {science}")
print("............................")

total=maths+english+science
print(f"total :{total}")
percentage=(total*100)/300
print(f"percentage{percentage}")
if percentage>=65:
    print("you have passed the exams")
else:
    print("you have fail the exams")
          
```
